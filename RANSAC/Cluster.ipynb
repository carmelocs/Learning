```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n_samples = 100
n_centers = 3

data = np.random.randint(0, 500, size=n_samples)
y = np.zeros_like(data)
centers = np.linspace(0, 500, num=n_centers)
print(data.shape, y.shape, centers)
```

```python
plt.scatter(data, y, color='yellowgreen', marker='.')
```

```python
# red, green, blue = np.zeros(n_samples)
dist_r, dist_g, dist_b = np.zeros(n_samples), np.zeros(n_samples), np.zeros(n_samples)
# dist_r, dist_g, dist_b = int(np.zeros(n_samples)), int(np.zeros(n_samples)), int(np.zeros(n_samples))
```

```python
for i in range(n_samples):
    dist_r[i] = (data[i] - centers[0])**2
    dist_g[i] = (data[i] - centers[1])**2
    dist_b[i] = (data[i] - centers[2])**2
```

```python
dist_r.shape
```

```python
red = (dist_r < dist_g) & (dist_r<dist_b)
green = (dist_g < dist_r) & (dist_g<dist_b)
blue = (dist_b < dist_r) & (dist_b<dist_r)
red_y = np.zeros_like(data[red])
green_y = np.zeros_like(data[green])
blue_y = np.zeros_like(data[blue])
```

```python
print(data[red].shape, red_y.shape)
```

```python
plt.scatter(data[red], red_y, color='red', marker='.')
plt.scatter(data[green], green_y, color='yellowgreen', marker='.')
plt.scatter(data[blue], blue_y, color='blue', marker='.')
```
