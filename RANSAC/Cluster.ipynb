```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
n_points = 500
n_samples = 100
n_centers = 3
n_iters = 1

np.random.seed(0)
data = np.random.randint(0, n_points, size=n_samples)
y = np.zeros_like(data)
centers = np.linspace(0, n_points, num=n_centers)
print(data.shape, y.shape, centers)

plt.scatter(data, y, color='yellowgreen', marker='.')
```

```python
dist_r, dist_g, dist_b = np.zeros(n_samples), np.zeros(n_samples), np.zeros(n_samples)
```

```python
for iter in range(n_iters):
    for i in range(n_samples):
        dist_r[i] = (data[i] - centers[0])**2
        dist_g[i] = (data[i] - centers[1])**2
        dist_b[i] = (data[i] - centers[2])**2

    red = (dist_r < dist_g) & (dist_r < dist_b)
    green = (dist_g < dist_r) & (dist_g < dist_b)
    blue = (dist_b < dist_r) & (dist_b < dist_g)

    centers[0] = (data[red].min() + data[red].max()) // 2
    centers[1] = (data[green].min() + data[green].max()) // 2
    centers[2] = (data[blue].min() + data[blue].max()) // 2
```
